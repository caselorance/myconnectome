This notebook summarizes QA metrics for the MyConnectome anatomical data. Initial QA processing was done on TACC.

```python
import os,glob
import pickle
import pandas as pd
import numpy
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

sns.set(style="whitegrid")

qadatadir='/Users/poldrack/Dropbox/data/selftracking/QA'

anatdatafile=os.path.join(qadatadir,'anat_qa.pkl')

anatdata=pickle.load(open(anatdatafile,'rb'))

subcode=anatdata.pop('subcode')
s=pd.DataFrame(anatdata,index=subcode)
```

```python
s=s.sort()
for var in ['mean_gm','mean_wm','std_bg']:
    s.drop(var,axis=1, inplace=True)
s.describe()
```
